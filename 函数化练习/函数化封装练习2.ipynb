## 🧠 背后原理：字典查找
在 Python 中，字典是一种键值对（key-value）结构。你可以通过键（key）来查找对应的值（value）。

```python
RNA_CODON_MAP = {
    "AUG": "Met",
    "GCU": "Ala",
    "AUU": "Ile",
    "GUA": "Val",
    "AAU": "Asn",
    "UAA": "STOP",
    "UAG": "STOP",
    "UGA": "STOP"
}
```
在这个字典中：

`"AUG"` 是一个键（key）

`"Met"` 是它对应的值（value）

✅ 为什么 RNA_CODON_MAP[codon] 成立
假设你已经从 RNA 序列中提取出一个密码子：

```python
codon = "AUG"
```
那么：

```python
amino_acid = RNA_CODON_MAP[codon]
```
就等价于：

```python
amino_acid = RNA_CODON_MAP["AUG"]  # 返回 "Met"
```
这条语句的作用是：

在字典 `RNA_CODON_MAP`中查找键 `"AUG"`

返回对应的值 `"Met"`

并将其赋值给变量 `amino_acid`

🛑 注意事项
如果你尝试查找一个不存在的键，比如：

```python
codon = "XYZ"
amino_acid = RNA_CODON_MAP[codon]  # 会报错：KeyError
```
所以在实际代码中，我们通常会先判断：

```python
if codon in RNA_CODON_MAP:
    amino_acid = RNA_CODON_MAP[codon]
```
或者使用 `.get()` 方法更安全：

```python
amino_acid = RNA_CODON_MAP.get(codon)
```
这样如果 `codon` 不存在，会返回 `None` 而不是报错。

🧪 编程题：DNA 密码子频率统计与可视化
背景：
DNA 序列由 A、T、G、C 组成，每 3 个碱基构成一个密码子。你需要统计每种密码子的出现次数，过滤掉只出现 1 次的密码子，并按频率从高到低排序，最后绘制柱状图展示结果。

🧩 任务要求
请完成以下步骤：

函数封装：定义函数 `count_codons(sequence)`，返回密码子频数字典。

过滤：只保留出现次数 ≥ 2 的密码子。

排序：按出现次数从高到低排序。

可视化：绘制柱状图，x 轴为密码子，y 轴为出现次数。

🧬 示例 DNA 序列
```python
dna_sequence = "ATGCGTGCTGCTTAAATGCGTAAACCCGGGTTTAAACCC"
```
🧠 提示
使用 `dict.get()` 进行计数

使用 `sorted()` 对字典按值排序

可视化推荐使用 `matplotlib.pyplot` 或 `seaborn`

````python
dna_sequence = "ATGCGTGCTGCTTAAATGCGTAAACCCGGGTTTAAACCC"
counts = {}
for i in range(0,len(dna_sequence),3):
    codon = dna_sequence[i:i+3]
    if len(codon) < 3:
        continue
    else:
        counts[codon] = counts.get(codon,0) +1
filtered_counts = {codon: count for codon, count in counts.items() if count >= 2}
# 🧠 步骤 1：把字典转成列表（键值对）

items = []
for codon in filtered_counts:
    count = filtered_counts[codon]
    items.append([codon, count])

# 🧠 步骤 2：手动排序（选择排序为例）
# 我们用选择排序（selection sort）来按 count 从大到小排序：


# 对 items 列表按 count 从大到小排序
for i in range(len(items)):
    max_index = i
    for j in range(i+1, len(items)):
        if items[j][1] > items[max_index][1]:
            max_index = j
    # 交换位置
    items[i], items[max_index] = items[max_index], items[i]

# 🧠 步骤 3：打印排序结果

for codon, count in items:
    print(f"{codon}: {count}")
````

````python
# ✅ 函数实现（纯手写逻辑）

def count_and_sort_codons(sequence):
    # 1. 统计密码子频率
    counts = {}
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        if len(codon) < 3:
            continue
        counts[codon] = counts.get(codon, 0) + 1

    # 2. 过滤掉只出现一次的密码子
    filtered = {}
    for codon in counts:
        if counts[codon] >= 2:
            filtered[codon] = counts[codon]

    # 3. 转换为列表形式 [codon, count]
    items = []
    for codon in filtered:
        items.append([codon, filtered[codon]])

    # 4. 使用选择排序按 count 从高到低排序
    for i in range(len(items)):
        max_index = i
        for j in range(i+1, len(items)):
            if items[j][1] > items[max_index][1]:
                max_index = j
        # 交换
        items[i], items[max_index] = items[max_index], items[i]

    # 5. 返回排序后的列表
    return items
````
