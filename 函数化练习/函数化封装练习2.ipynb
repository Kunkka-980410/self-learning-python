````python
# ‚úÖ ÂáΩÊï∞ÂÆûÁé∞ÔºàÁ∫ØÊâãÂÜôÈÄªËæëÔºâ

def count_and_sort_codons(sequence):
    # 1. ÁªüËÆ°ÂØÜÁ†ÅÂ≠êÈ¢ëÁéá
    counts = {}
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        if len(codon) < 3:
            continue
        counts[codon] = counts.get(codon, 0) + 1

    # 2. ËøáÊª§ÊéâÂè™Âá∫Áé∞‰∏ÄÊ¨°ÁöÑÂØÜÁ†ÅÂ≠ê
    filtered = {}
    for codon in counts:
        if counts[codon] >= 2:
            filtered[codon] = counts[codon]

    # 3. ËΩ¨Êç¢‰∏∫ÂàóË°®ÂΩ¢Âºè [codon, count]
    items = []
    for codon in filtered:
        items.append([codon, filtered[codon]])

    # 4. ‰ΩøÁî®ÈÄâÊã©ÊéíÂ∫èÊåâ count ‰ªéÈ´òÂà∞‰ΩéÊéíÂ∫è
    for i in range(len(items)):
        max_index = i
        for j in range(i+1, len(items)):
            if items[j][1] > items[max_index][1]:
                max_index = j
        # ‰∫§Êç¢
        items[i], items[max_index] = items[max_index], items[i]

    # 5. ËøîÂõûÊéíÂ∫èÂêéÁöÑÂàóË°®
    return items
````

````python
import matplotlib.pyplot as plt

dna_sequence = "ATGCGTGCTGCTTAAATGCGTAAACCCGGGTTTAAACCC"
def analyze_frames_and_plot(sequence):
    """
    ÂàÜÊûê DNA Â∫èÂàóÂú®‰∏â‰∏™ÈòÖËØªÊ°Ü‰∏≠ÁöÑÂØÜÁ†ÅÂ≠êÈ¢ëÁéáÔºåÂπ∂ÁªòÂà∂ËøáÊª§ÂêéÁöÑÊü±Áä∂Âõæ„ÄÇ
    
    ÂèÇÊï∞Ôºö
        sequence (str): DNA Â∫èÂàóÂ≠óÁ¨¶‰∏≤
    """
    for frame in range(3):
        counts = {}
        for i in range(frame, len(sequence), 3):
            codon = sequence[i:i+3]
            if len(codon) < 3:
                continue
            counts[codon] = counts.get(codon, 0) + 1

        # ËøáÊª§È¢ëÁéáÂ§ß‰∫é 1 ÁöÑÂØÜÁ†ÅÂ≠ê
        filtered_counts = {codon: count for codon, count in counts.items() if count >= 1}

        # Â¶ÇÊûúÊ≤°ÊúâÂØÜÁ†ÅÂ≠êÊª°Ë∂≥Êù°‰ª∂ÔºåË∑≥ËøáÁªòÂõæ
        if not filtered_counts:
            print(f"Frame {frame}: Ê≤°ÊúâÈ¢ëÁéá ‚â• 1 ÁöÑÂØÜÁ†ÅÂ≠êÔºåË∑≥ËøáÁªòÂõæ")
            continue

        # ÁªòÂà∂Êü±Áä∂Âõæ
        sorted_codons = sorted(filtered_counts.keys())
        sorted_counts = [filtered_counts[codon] for codon in sorted_codons]

        plt.figure(figsize=(8, 5))
        bars = plt.bar(sorted_codons, sorted_counts, color='skyblue', edgecolor='black')

        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, str(height),
                     ha='center', va='bottom', fontsize=10)

        plt.title(f"Frame {frame}ÔºöËøáÊª§ÂêéÂØÜÁ†ÅÂ≠êÈ¢ëÁéáÊü±Áä∂Âõæ", fontsize=14)
        plt.xlabel("ÂØÜÁ†ÅÂ≠ê", fontsize=12)
        plt.ylabel("Âá∫Áé∞Ê¨°Êï∞", fontsize=12)
        plt.ylim(0, max(sorted_counts) + 1)
        plt.tight_layout()
        plt.show()
analyze_frames_and_plot('ATG')
````
