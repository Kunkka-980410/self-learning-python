```python
'''
计算每个基因的组内均值（所有样本的平均表达值）

按照是否高于该均值，把样本分成高表达组和低表达组

分别计算高表达组和低表达组的平均值

计算两组均值的差值（组件均值差）

把这些信息存入一个新字典 expression_report

'''

expression_data = {
    "ENSG00000141510": {
        "gene_name": "TP53",
        "sample1": 50,
        "sample2": 60,
        "sample3": 55,
        "sample4": 40,
        "sample5": 30,
        "sample6": 70
    },
    "ENSG00000012048": {
        "gene_name": "BRCA1",
        "sample1": 5,
        "sample2": 3,
        "sample3": 4,
        "sample4": 6,
        "sample5": 2,
        "sample6": 8
    }
}



'''
伪代码思路

1.先设置一个空字典 expression_report={}，
可以假设下内部形式 expression_report={“基因名”：{“表达值均值”；“表达值均值”，
                                                “高表达组”：“样本1，样本2，样本3”，
                                                “低表达组”：“样本4，样本5，样本6”，
                                                “高表达组平均值”：“平均值”，
                                                “低表达组平均值”：“平均值”，
                                                “组间差值”：“组间差值”}}


2.遍历原字典的每个键值对   
for k,v in expression_data.items(): 

3.取出基因名，作为新字典expression_report={}的一个键
expression_report["gene_name"] = v["gene_name"]

4.建2个存sample表达值的空字典，一个是高表达组，一个是低表达组

5.用一个变量来遍历第二步取出的v.items()，

6.分别求出两个基因的sample值的 mean 

7.将两个 mean 值作为新键值对添加到字典 expression_report={} 中

8.分别用基因的 mean 值来比较 sample 表达值，高于表达值放入高表达组，低于放入低表达组

9.作为新键值对添加到 expression_report={} 中

10.分布计算2个基因的高表达组和低表达组的均值

11.2个基因的高表达组和低表达组均值作为键值对添加到 expression_report={} 中

12.分别计算2个基因的高低表达组均值的差异

13.作为键值对添加到 expression_report={} 中

14.在字典中添加新的键值对

'''
  
```

```python
import math # 导入math用于检查nan值，虽然这里用不到，但保留代码习惯

expression_data = {
    "ENSG00000141510": {
        "gene_name": "TP53",
        "sample1": 50,
        "sample2": 60,
        "sample3": 55,
        "sample4": 40,
        "sample5": 30,
        "sample6": 70
    },
    "ENSG00000012048": {
        "gene_name": "BRCA1",
        "sample1": 5,
        "sample2": 3,
        "sample3": 4,
        "sample4": 6,
        "sample5": 2,
        "sample6": 8
    }
}

# 1. 初始化结果字典
expression_report = {}

# 2. 遍历原字典的每个键值对 (k: 基因ID, v: 基因数据)
for gene_id, gene_data in expression_data.items():
    
    # 获取所有样本的表达值 (排除 'gene_name')
    # 使用字典推导式快速过滤出样本数据
    sample_values = {
        k: v for k, v in gene_data.items() if k.startswith('sample')
    }
    
    # 提取表达值列表用于计算
    values_list = list(sample_values.values())
    
    # --- 任务 1: 计算组内均值 ---
    group_mean = sum(values_list) / len(values_list) if values_list else 0
    
    # 3. 初始化当前基因的报告字典
    gene_report = {
        "gene_name": gene_data["gene_name"],
        "overall_mean": round(group_mean, 4)
    }

    # --- 任务 2: 划分高表达组和低表达组 ---
    high_group_names = []
    low_group_names = []
    high_group_values = []
    low_group_values = []
    
    # 5. 遍历样本
    for sample_name, value in sample_values.items():
        if value > group_mean:
            high_group_names.append(sample_name)
            high_group_values.append(value)
        else: # 低于或等于均值都归为低表达组
            low_group_names.append(sample_name)
            low_group_values.append(value)
    
    # --- 任务 3: 分别计算高表达组和低表达组的平均值 ---
    
    # 如果列表为空，则均值记为 None 或 0
    high_mean = sum(high_group_values) / len(high_group_values) if high_group_values else None
    low_mean = sum(low_group_values) / len(low_group_values) if low_group_values else None
    
    # --- 任务 4: 计算两组均值的差值 (高表达组均值 - 低表达组均值) ---
    mean_difference = round(high_mean - low_mean, 4) if high_mean is not None and low_mean is not None else None
    
    # 7, 9, 11, 13. 将所有信息添加到当前基因的报告字典中
    gene_report["high_group_samples"] = high_group_names
    gene_report["low_group_samples"] = low_group_names
    gene_report["high_group_mean"] = round(high_mean, 4) if high_mean is not None else "N/A"
    gene_report["low_group_mean"] = round(low_mean, 4) if low_mean is not None else "N/A"
    gene_report["mean_difference"] = mean_difference if mean_difference is not None else "N/A"
    
    # 14. 将当前基因的报告添加到最终结果字典
    expression_report[gene_id] = gene_report

# 打印最终结果字典
import json
print(json.dumps(expression_report, indent=4))
```
