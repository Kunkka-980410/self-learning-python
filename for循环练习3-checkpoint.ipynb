```python
'''
è®¡ç®—æ¯ä¸ªåŸºå› çš„ç»„å†…å‡å€¼ï¼ˆæ‰€æœ‰æ ·æœ¬çš„å¹³å‡è¡¨è¾¾å€¼ï¼‰

æŒ‰ç…§æ˜¯å¦é«˜äºè¯¥å‡å€¼ï¼ŒæŠŠæ ·æœ¬åˆ†æˆé«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„

åˆ†åˆ«è®¡ç®—é«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„çš„å¹³å‡å€¼

è®¡ç®—ä¸¤ç»„å‡å€¼çš„å·®å€¼ï¼ˆç»„ä»¶å‡å€¼å·®ï¼‰

æŠŠè¿™äº›ä¿¡æ¯å­˜å…¥ä¸€ä¸ªæ–°å­—å…¸ expression_report

'''

expression_data = {
    "ENSG00000141510": {
        "gene_name": "TP53",
        "sample1": 50,
        "sample2": 60,
        "sample3": 55,
        "sample4": 40,
        "sample5": 30,
        "sample6": 70
    },
    "ENSG00000012048": {
        "gene_name": "BRCA1",
        "sample1": 5,
        "sample2": 3,
        "sample3": 4,
        "sample4": 6,
        "sample5": 2,
        "sample6": 8
    }
}



'''
ä¼ªä»£ç æ€è·¯

1.å…ˆè®¾ç½®ä¸€ä¸ªç©ºå­—å…¸ expression_report={}ï¼Œ
å¯ä»¥å‡è®¾ä¸‹å†…éƒ¨å½¢å¼ expression_report={â€œåŸºå› åâ€ï¼š{â€œè¡¨è¾¾å€¼å‡å€¼â€ï¼›â€œè¡¨è¾¾å€¼å‡å€¼â€ï¼Œ
                                                â€œé«˜è¡¨è¾¾ç»„â€ï¼šâ€œæ ·æœ¬1ï¼Œæ ·æœ¬2ï¼Œæ ·æœ¬3â€ï¼Œ
                                                â€œä½è¡¨è¾¾ç»„â€ï¼šâ€œæ ·æœ¬4ï¼Œæ ·æœ¬5ï¼Œæ ·æœ¬6â€ï¼Œ
                                                â€œé«˜è¡¨è¾¾ç»„å¹³å‡å€¼â€ï¼šâ€œå¹³å‡å€¼â€ï¼Œ
                                                â€œä½è¡¨è¾¾ç»„å¹³å‡å€¼â€ï¼šâ€œå¹³å‡å€¼â€ï¼Œ
                                                â€œç»„é—´å·®å€¼â€ï¼šâ€œç»„é—´å·®å€¼â€}}


2.éå†åŸå­—å…¸çš„æ¯ä¸ªé”®å€¼å¯¹   
for k,v in expression_data.items(): 

3.å–å‡ºåŸºå› åï¼Œä½œä¸ºæ–°å­—å…¸expression_report={}çš„ä¸€ä¸ªé”®
expression_report["gene_name"] = v["gene_name"]

4.å»º2ä¸ªå­˜sampleè¡¨è¾¾å€¼çš„ç©ºå­—å…¸ï¼Œä¸€ä¸ªæ˜¯é«˜è¡¨è¾¾ç»„ï¼Œä¸€ä¸ªæ˜¯ä½è¡¨è¾¾ç»„

5.ç”¨ä¸€ä¸ªå˜é‡æ¥éå†ç¬¬äºŒæ­¥å–å‡ºçš„v.items()ï¼Œ

6.åˆ†åˆ«æ±‚å‡ºä¸¤ä¸ªåŸºå› çš„sampleå€¼çš„ mean 

7.å°†ä¸¤ä¸ª mean å€¼ä½œä¸ºæ–°é”®å€¼å¯¹æ·»åŠ åˆ°å­—å…¸ expression_report={} ä¸­

8.åˆ†åˆ«ç”¨åŸºå› çš„ mean å€¼æ¥æ¯”è¾ƒ sample è¡¨è¾¾å€¼ï¼Œé«˜äºè¡¨è¾¾å€¼æ”¾å…¥é«˜è¡¨è¾¾ç»„ï¼Œä½äºæ”¾å…¥ä½è¡¨è¾¾ç»„

9.ä½œä¸ºæ–°é”®å€¼å¯¹æ·»åŠ åˆ° expression_report={} ä¸­

10.åˆ†å¸ƒè®¡ç®—2ä¸ªåŸºå› çš„é«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„çš„å‡å€¼

11.2ä¸ªåŸºå› çš„é«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„å‡å€¼ä½œä¸ºé”®å€¼å¯¹æ·»åŠ åˆ° expression_report={} ä¸­

12.åˆ†åˆ«è®¡ç®—2ä¸ªåŸºå› çš„é«˜ä½è¡¨è¾¾ç»„å‡å€¼çš„å·®å¼‚

13.ä½œä¸ºé”®å€¼å¯¹æ·»åŠ åˆ° expression_report={} ä¸­

14.åœ¨å­—å…¸ä¸­æ·»åŠ æ–°çš„é”®å€¼å¯¹

'''
  
```

```python
import math # å¯¼å…¥mathç”¨äºæ£€æŸ¥nanå€¼ï¼Œè™½ç„¶è¿™é‡Œç”¨ä¸åˆ°ï¼Œä½†ä¿ç•™ä»£ç ä¹ æƒ¯

expression_data = {
    "ENSG00000141510": {
        "gene_name": "TP53",
        "sample1": 50,
        "sample2": 60,
        "sample3": 55,
        "sample4": 40,
        "sample5": 30,
        "sample6": 70
    },
    "ENSG00000012048": {
        "gene_name": "BRCA1",
        "sample1": 5,
        "sample2": 3,
        "sample3": 4,
        "sample4": 6,
        "sample5": 2,
        "sample6": 8
    }
}

# 1. åˆå§‹åŒ–ç»“æœå­—å…¸
expression_report = {}

# 2. éå†åŸå­—å…¸çš„æ¯ä¸ªé”®å€¼å¯¹ (k: åŸºå› ID, v: åŸºå› æ•°æ®)
for gene_id, gene_data in expression_data.items():
    
    # è·å–æ‰€æœ‰æ ·æœ¬çš„è¡¨è¾¾å€¼ (æ’é™¤ 'gene_name')
    # ä½¿ç”¨å­—å…¸æ¨å¯¼å¼å¿«é€Ÿè¿‡æ»¤å‡ºæ ·æœ¬æ•°æ®
    sample_values = {
        k: v for k, v in gene_data.items() if k.startswith('sample')
    }
    
    # æå–è¡¨è¾¾å€¼åˆ—è¡¨ç”¨äºè®¡ç®—
    values_list = list(sample_values.values())
    
    # --- ä»»åŠ¡ 1: è®¡ç®—ç»„å†…å‡å€¼ ---
    group_mean = sum(values_list) / len(values_list) if values_list else 0
    
    # 3. åˆå§‹åŒ–å½“å‰åŸºå› çš„æŠ¥å‘Šå­—å…¸
    gene_report = {
        "gene_name": gene_data["gene_name"],
        "overall_mean": round(group_mean, 4)
    }

    # --- ä»»åŠ¡ 2: åˆ’åˆ†é«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„ ---
    high_group_names = []
    low_group_names = []
    high_group_values = []
    low_group_values = []
    
    # 5. éå†æ ·æœ¬
    for sample_name, value in sample_values.items():
        if value > group_mean:
            high_group_names.append(sample_name)
            high_group_values.append(value)
        else: # ä½äºæˆ–ç­‰äºå‡å€¼éƒ½å½’ä¸ºä½è¡¨è¾¾ç»„
            low_group_names.append(sample_name)
            low_group_values.append(value)
    
    # --- ä»»åŠ¡ 3: åˆ†åˆ«è®¡ç®—é«˜è¡¨è¾¾ç»„å’Œä½è¡¨è¾¾ç»„çš„å¹³å‡å€¼ ---
    
    # å¦‚æœåˆ—è¡¨ä¸ºç©ºï¼Œåˆ™å‡å€¼è®°ä¸º None æˆ– 0
    high_mean = sum(high_group_values) / len(high_group_values) if high_group_values else None
    low_mean = sum(low_group_values) / len(low_group_values) if low_group_values else None
    
    # --- ä»»åŠ¡ 4: è®¡ç®—ä¸¤ç»„å‡å€¼çš„å·®å€¼ (é«˜è¡¨è¾¾ç»„å‡å€¼ - ä½è¡¨è¾¾ç»„å‡å€¼) ---
    mean_difference = round(high_mean - low_mean, 4) if high_mean is not None and low_mean is not None else None
    
    # 7, 9, 11, 13. å°†æ‰€æœ‰ä¿¡æ¯æ·»åŠ åˆ°å½“å‰åŸºå› çš„æŠ¥å‘Šå­—å…¸ä¸­
    gene_report["high_group_samples"] = high_group_names
    gene_report["low_group_samples"] = low_group_names
    gene_report["high_group_mean"] = round(high_mean, 4) if high_mean is not None else "N/A"
    gene_report["low_group_mean"] = round(low_mean, 4) if low_mean is not None else "N/A"
    gene_report["mean_difference"] = mean_difference if mean_difference is not None else "N/A"
    
    # 14. å°†å½“å‰åŸºå› çš„æŠ¥å‘Šæ·»åŠ åˆ°æœ€ç»ˆç»“æœå­—å…¸
    expression_report[gene_id] = gene_report

# æ‰“å°æœ€ç»ˆç»“æœå­—å…¸
import json
print(json.dumps(expression_report, indent=4))
```

```python
import math # å¯¼å…¥ math åº“ç”¨äºæ£€æŸ¥æ½œåœ¨çš„ NaN/None å€¼ï¼Œè™½ç„¶åœ¨æœ¬ä¾‹ä¸­ä¸ä¼šå‘ç”Ÿ

# ä¿®æ­£åçš„åŸå§‹æ•°æ®
expression_data = {
    "TP53":{
        "gene_name":"TP53",
        "sample1":50,
        "sample2":60,
        "sample3":55,
        "sample4":40,
        "sample5":30,
        "sample6":70
    },
    "BRCA1":{
        "gene_name":"BRCA1",
        "sample1":5,
        "sample2":3,
        "sample3":4,
        "sample4":6,
        "sample5":2,
        "sample6":8
    }
}

# æœ€ç»ˆç»“æœå­—å…¸
expression_shift = {}

# ã€ç¬¬ä¸€å±‚å¾ªç¯ã€‘ï¼šéå†æ¯ä¸€ä¸ªåŸºå› 
for gene_id, info in expression_data.items():
    
    # ----------------------------------------------------
    # æ­¥éª¤ 1ï¼šå‡†å¤‡æ•°æ®å’Œè®¡ç®—ç»„å†…å‡å€¼
    # ----------------------------------------------------
    
    sample_values_list = []
    
    # å†…éƒ¨å¾ªç¯ï¼šæå–æ‰€æœ‰æ ·æœ¬å€¼
    for k, v in info.items():
        if k.startswith("sample"):
            sample_values_list.append(v)
            
    # è®¡ç®—å‡å€¼
    if sample_values_list:
        overall_mean = sum(sample_values_list) / len(sample_values_list)
    else:
        overall_mean = 0

    # ----------------------------------------------------
    # æ­¥éª¤ 2ï¼šæ‰¾å‡ºé«˜äº/ä½äºå‡å€¼çš„æ ·æœ¬ï¼Œå¹¶å­˜å…¥æ–°çš„å­—å…¸
    # ----------------------------------------------------
    
    high_samples = {}
    low_samples = {}
    high_group_values = []
    low_group_values = []
    
    # å†…éƒ¨å¾ªç¯ï¼šéå†æ ·æœ¬ï¼Œè¿›è¡Œåˆ’åˆ†
    for k, v in info.items():
        if k.startswith("sample"):
            if v > overall_mean:
                # é«˜è¡¨è¾¾ç»„ (å€¼ä¸¥æ ¼å¤§äºå‡å€¼)
                high_samples[k] = v
                high_group_values.append(v)
            else:
                # ä½è¡¨è¾¾ç»„ (å€¼å°äºæˆ–ç­‰äºå‡å€¼)
                low_samples[k] = v
                low_group_values.append(v)

    
    # ----------------------------------------------------
    # æ­¥éª¤ 3ï¼šè®¡ç®—é«˜/ä½ç»„å‡å€¼
    # ----------------------------------------------------
    
    # è®¡ç®—é«˜è¡¨è¾¾ç»„å‡å€¼
    if high_group_values:
        high_mean = sum(high_group_values) / len(high_group_values)
    else:
        high_mean = "N/A"
        
    # è®¡ç®—ä½è¡¨è¾¾ç»„å‡å€¼
    if low_group_values:
        low_mean = sum(low_group_values) / len(low_group_values)
    else:
        # å¦‚æœæ‰€æœ‰æ ·æœ¬éƒ½é«˜äºå‡å€¼ï¼Œlow_group_values å¯èƒ½ä¸ºç©º
        low_mean = "N/A" 
        
    # ----------------------------------------------------
    # æ­¥éª¤ 4ï¼šè®¡ç®—ç»„å†…é«˜ä½è¡¨è¾¾ç»„å‡å€¼ä¹‹å·® (Shift)
    # ----------------------------------------------------
    
    if high_mean != "N/A" and low_mean != "N/A":
        shift_value = high_mean - low_mean
    else:
        shift_value = "N/A"

    
    # ----------------------------------------------------
    # æ­¥éª¤ 5ï¼šå°†ç»“æœå­˜å…¥ expression_shift å­—å…¸
    # ----------------------------------------------------
    
    # æŒ‰ç…§è¦æ±‚çš„æ ¼å¼æ•´ç†æ•°æ®
    report = {
        "overall_mean": round(overall_mean, 4),
        "high_samples": high_samples,
        "low_samples": low_samples,
        "high_mean": round(high_mean, 4) if high_mean != "N/A" else "N/A",
        "low_mean": round(low_mean, 4) if low_mean != "N/A" else "N/A",
        "shift": round(shift_value, 4) if shift_value != "N/A" else "N/A"
    }
    
    expression_shift[gene_id] = report


# æ‰“å°æœ€ç»ˆç»“æœï¼ˆä½¿ç”¨ for å¾ªç¯æ‰“å°å­—å…¸å†…å®¹ï¼‰
print("--- æœ€ç»ˆè¡¨è¾¾åç§»æŠ¥å‘Š (expression_shift) ---")

for gene_id, report in expression_shift.items():
    print(f"\n[{gene_id}]:")
    print(f"  æ€»ä½“å‡å€¼ (overall_mean): {report['overall_mean']}")
    print(f"  é«˜è¡¨è¾¾ç»„ (high_mean: {report['high_mean']}): {report['high_samples']}")
    print(f"  ä½è¡¨è¾¾ç»„ (low_mean: {report['low_mean']}): {report['low_samples']}")
    print(f"  ç»„å†…åç§» (shift): {report['shift']}")
```
