```python
expression_data = {
    "ENSG00000141510": {
        "gene_name": "TP53",
        "sample1": 50,
        "sample2": 60,
        "sample3": 55
    },
    "ENSG00000012048": {
        "gene_name": "BRCA1",
        "sample1": 5,
        "sample2": 3,
        "sample3": 4
    },
    "ENSG00000139618": {
        "gene_name": "BRCA2",
        "sample1": 100,
        "sample2": 120,
        "sample3": 110
    },
    "ENSG00000210082": {
        "gene_name": "MIR21",
        "sample1": 0,
        "sample2": 1,
        "sample3": 0
    }
}


report = {}

for gene_id, info in expression_data.items():
    gene_name = info["gene_name"]
    high_sample_values = {
    sample: value
    for sample, value in info.items()
    if sample != "gene_name" and value > 50
}
    # 只有在有高表达样本时才添加
    if high_sample_values:
        report[gene_name] = high_sample_values


```

```python
'''
if sample != "gene_name" and value > 50这个操作是什么意思
是在一个 for 循环中，用来筛选表达量高于 50 的样本，同时排除掉不是样本的字段（比如 gene_name）。我们来逐步解释它的含义：

✅ 背景：你正在遍历一个嵌套字典
比如：

python
info = {
    "gene_name": "TP53",
    "sample1": 50,
    "sample2": 60,
    "sample3": 55
}
你用：

python
for sample, value in info.items():
会依次得到：

sample = "gene_name", value = "TP53"

sample = "sample1", value = 50

sample = "sample2", value = 60

sample = "sample3", value = 55

✅ 条件解释：
python
if sample != "gene_name" and value > 50
🔹 sample != "gene_name"
排除掉 "gene_name" 这个字段，因为它不是表达量，只是个标签。

🔹 value > 50
只保留表达量大于 50 的样本。

✅ 合起来的意思：
“如果当前键不是 gene_name，并且它的表达量大于 50，就把它当作高表达样本处理。

🧠 举个例子：
在 TP53 的表达数据中：

python
"sample1": 50   → 不满足 value > 50  
"sample2": 60   → 满足条件 ✅  
"sample3": 55   → 满足条件 ✅  
"gene_name": "TP53" → 被排除 ❌
最终你会得到：

python
["sample2", "sample3"]

'''
```
