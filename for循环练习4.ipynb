````python
stable_genes = {}

for gene, info in expression_data.items():
    gene_name = info["gene_name"]
    values = []

    # 收集样本表达值
    for k, v in info.items():
        if k != "gene_name":
            values.append(v)

    # 计算均值
    mean = sum(values) / len(values)

    # 计算标准差（展开为完整 for 循环）
    squared_diffs = []
    for x in values:
        diff = x - mean
        squared = diff ** 2
        squared_diffs.append(squared)

    variance = sum(squared_diffs) / len(values)
    std = variance ** 0.5

    # 计算最大值与最小值之差
    diff = max(values) - min(values)

    # 判断表达稳定性
    if std <= 5 and diff <= 10:
        stable_genes[gene_name] = {
            "mean": round(mean, 2),
            "std": round(std, 2),
            "diff": diff
        }
 
````
