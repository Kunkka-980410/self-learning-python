````python
def analyze_expression_groups(expression_data, treatment_samples, control_samples,target_gene = None):
    group_info = {}

    for gene, info in expression_data.items():
        gene_name = gene
        treatment = {}
        control = {}
         # 如果指定了 target_gene，只处理它
        if target_gene and gene != target_gene:
            continue
        
        for sample, value in info.items():
            if sample in treatment_samples:
                treatment[sample] = value
            elif sample in control_samples:
                control[sample] = value

        treatment_mean = sum(treatment.values()) / len(treatment.values())
        control_mean = sum(control.values()) / len(control.values())

        if treatment_mean > control_mean:
            status = "high"
        elif treatment_mean < control_mean:
            status = "low"
        else:
            status = "no change"

        group_info[gene] = {
            "treatment": treatment,
            "control": control,
            "treatment_mean": round(treatment_mean, 2),
            "control_mean": round(control_mean, 2),
            "status": status
        }

    return group_info


# ✅ 在函数外定义分组
treatment_samples = ["sample_1", "sample_2"]
control_samples = ["sample_3", "sample_4"] 

analyze_expression_groups(expression_data, treatment_samples, control_samples,target_gene = 'BRCA1')
````

✅ 在普通的 for 循环里：
你写的是脚本式代码，所有变量都是全局变量，比如：

python
for gene, info in expression_data.items():
    gene_name = gene
这里你把 gene_name 赋值成 gene，没问题，因为你没有用 gene_name 作为参数，也没有函数作用域限制。

❌ 在函数里：
你写的是：

python
def high_expression_mean(gene_name):
    for gene, info in expression_data.items():
        gene_name = gene  # ❌ 这里覆盖了你传入的参数
这就出事了，因为：

gene_name 是你传入的参数，比如 "TP53"
